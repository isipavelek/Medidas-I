### Punto 20mAs

### Tipo A

```python
import numpy as np
import statistics

mAsP=[20.3,19,19.5,20.1,19.4,20,18.3,20.3,20.2,19.8]
mAsUTN=[20.35,19.06,19.71,20.34,19.69,20.03,18.39,20.37,20.38,20.02]
N=len(mAsP)
mAsPm=np.mean(mAsP)
mAsUTNm=np.mean(mAsUTN)
SmAsP=statistics.stdev(mAsP)
SmAsUTN=statistics.stdev(mAsUTN)
uamAsPr=SmAsP/np.sqrt(N)
uamAsUTNr=SmAsUTN/np.sqrt(N)
print('mAsPm=',mAsPm,'mAs') 
print('mAsUTNm=',mAsUTNm,'mAs') 
print('S(mAsP)=',SmAsP,'mAs') 
print('S(mAsUTN)=',SmAsUTN,'mAs') 
print('uar(mAsP)=',uamAsPr,'mAs') 
print('uar(mAsUTN)=',uamAsUTNr,'mAs')
print("    ")
print("TIPO B")
DELTAmAUTNr=(0.1/mAsUTNm)
DELTAtUTNr=(10/10000000)
DeltamAsUTNr=DELTAmAUTNr+DELTAtUTNr
ubrmAsUTNr=((DeltamAsUTNr*mAsUTNm)/np.sqrt(3))
print('mAs UTN ubr(mAs)=',ubrmAsUTNr,'mAs') 
## mAsPatron
DeltamAsP=(2/100)*mAsPm
ubrmAsP=(DeltamAsP/np.sqrt(3))
print('Patron: ubr(mAs)=',ubrmAsP,'mAs') 

print("   ")
print("Combinando")
ucmAsUTN=np.sqrt(uamAsUTNr**2+ubrmAsUTNr**2)
ucmAsP=np.sqrt(ubrmAsP**2+uamAsPr**2)
URmAsUTN=2*ucmAsUTN
URmAsP=2*ucmAsP
print('Up(mAs)=',URmAsP,'mAs')
print('U_UTN(mAs)=',URmAsUTN,'mAs')
print("     ")
print("Resultado de la medición:")
print("     ")
  
print(f"U_patron(mAs)=({mAsPm:.2f}±{URmAsP.round(2):.2f})mAs")
print("K=2 con el 95,45% de nivel de confianza")
print("     ")  
print(f"U_UTN(mAs)=({mAsUTNm:.1f}±{URmAsUTN.round(2):.2f})mAs")
print("K=2 con el 95,45% de nivel de confianza")
```
