## Calculo del valor eficaz

Para la medición de tension vemos que la señal excursiona por 6 divisiones del osciloscopio por lo que tenemos:

$Vpp=6 Div \cdot \frac{1V}{Div}= 6V$

$\Delta Vpp=(\frac{2 \cdot 6V}{100} + 0,1div \cdot \frac{1V}{div} + 1mV)= 221mv $

$\mu_R(Vpp)= \frac{\Delta Vpp}{\sqrt{3} \cdot 6V}= 0,0212657 $

$\mu_R(FP)= \frac{0,5}{100 \cdot \sqrt{3}}= 0,0028867 $

$V_{RMS}=\frac{Vpp}{\sqrt{2}} \cdot 10 =42,4264V $

$\mu_R (V_{RMS})=\sqrt{\mu_R^2(Vpp) + \mu_R^2(FP)} =0,021460766$

$\mu (V_{RMS}) = V_{RMS} \cdot \mu_R (V_{RMS}) = 0,12876460174 V $

$U(V_{RMS})= K \cdot \mu(V_{RMS})= 0,257529203481V$

$V_{RMS}=(42,42±0,26) V$

## Calculo de la frecuencia con el osciloscopio

$T=2,6 Div \cdot \frac{2ms}{Div}= 5,2ms$

$\Delta T=(\frac{0,5 \cdot 5,2ms}{100} + 0,1div \cdot \frac{2ms}{div})= 0,226ms $

$\mu_R(T)=\frac{\Delta T}{\sqrt{3}\cdot 5,2ms}=0,0250925309302$

$f=\frac{1}{T}=192,307692308Hz$

$\mu_R(T)=\mu_R(f)$

$\mu(f)=\mu_R(f) \cdot 192,307692308Hz=4,82548671916Hz$

$U(F)=K \cdot \mu(f)=9,65097343832 Hz$

$f=(192,3±9,7)Hz$

## Calculo de la frecuencia con frecuencimetro  en modo medicion de periodo promediado x 10

Estabilidad de la base de tiempo:

$\frac{\Delta f_{BT}}{f_{BT}}=5ppm+0,1\frac{ppm}{mes} \cdot 12 meses =6,2ppm$

Incertidumbre de cuantización:

$\frac{\Delta N}{N}=\frac{1}{519582}=1,924624ppm$

Error debido al disparo:

$V(t)=V_SPsin(\omega t)$ 

$\Delta V=V_{NP}$

$SNR=\frac{V_{SP}}{V_{NP}}$

$\Delta_T=\frac{\Delta V}{[\frac{dV}{dt}]_{t=t_{TRG}}}$

Sabemos que el disparo se produce cuando la señal está en $ \frac{V_{SP}}{2} $

$\frac{V_{SP}}{2}=V_{SP}sin(\omega t)$

$arcsin(\frac{1}{2})=\omega t$

$\frac{\pi}{\omega 6}=t$

$\Delta_T=\frac{V_{NP}}{[V_{SP}\cdot \omega \cdot cos(\omega 
t)]_{t=\frac{\pi}{6\omega}}}$

$\Delta_T=\frac{V_{NP}}{V_{SP}\cdot \omega \cdot \frac{\sqrt{3}}{2}}$

$\frac{\Delta_T}{T}=\frac{V_{NP}}{V_{SP}\sqrt3 \pi}$

$e_D=2\cdot \frac{\Delta_T}{T}$

$e_D=\frac{2}{\sqrt3 \pi SNR}$


Reemplazando:

$e_D=\frac{2}{\sqrt3 \pi 1000}=0,36755 \cdot 10^{-3}$

$\frac{\Delta_T}{T}=±(\frac{\Delta f_{BT}}{f_{BT}}+\frac{\Delta N}{N}+\frac{e_D}{M})= 375,6748 ppm$

M=1 (promediado por 10)

$\mu_R(T)=\frac{\frac{\Delta_T}{T}}{sqrt3}=216,8959 ppm$

$\mu_R(T)=\mu_R(f)$

$T=\frac{N}{F_{BT} \cdot 10^1}=5,19582 ms$

$f=\frac1T= 192,4624247Hz$

$\mu_(f)=\mu_R(f)\cdot f=0,041744 Hz $

$U(f)=k \mu_(f)=0,083488 Hz$

$f=(192,462±0,083) Hz$

## Calculo de la frecuencia con frecuencimetro  en modo frecuencimetro con gate 10s

Del punto C sabemos:

$f_x=\frac{10^M F_{BT}}{N}$

$f_x=\frac{10 10MHz}{519582}$

Ahora como estamos en modo frecuencia con gate de 10s sabemos que:

$f_x=\frac{N}{10s}$

Igualando y despejando:

$N=\frac{10s \cdot 10\cdot 10MHz}{519582}=1924,6240$

Como N es un contador descartamos los decimales y nos queda que:

N=1924

En este caso ahora 

$f_x=\frac{N}{10s}$

$f_x= 192,4 Hz$

El error de la estabilidad de la base de tiempo es el mismo, no tenemos error de disparo, y vamos a calcular el error de cuantización.

$\frac{\Delta N}{N}=\frac{1}{1924}=519,75 ppm$

$\frac{\Delta_f}{f}=±(\frac{\Delta f_{BT}}{f_{BT}}+\frac{\Delta N}{N})= 525,95 ppm$

$\mu(f)=\frac{ 525,95 ppm x 192,4 Hz}{sqrt3 }=0,058423 Hz$

$U(f)=K \cdot \mu(f)= 0,116847Hz$

f=(192,40±0,12) Hz

Calculo de la incertidumbre:

```python
import numpy as np
import statistics as st


Freqs=np.array([800013,800011,800014,800011,800010])
Freqsm=np.mean(Freqs)
SFreqs=st.stdev(Freqs)
uaFreqs=SFreqs/np.sqrt(len(Freqs))
print(Freqsm)
print(SFreqs)
print(uaFreqs)
```
